```python
import sys
from pathlib import Path
sys.path.append('..')

import matplotlib.pyplot as plt
import numpy as np
from babel_ai.analyzer import SimilarityAnalyzer
```

```python
def extract_messages(conversation):
    """Extract just the message content from a conversation list."""
    return [msg.split(': ')[1] if ': ' in msg else msg for msg in conversation]

def analyze_conversation(conversation, analyzer):
    """Analyze a conversation and return metrics."""
    messages = extract_messages(conversation)
    metrics = []
    
    for i in range(len(messages)):
        analysis = analyzer.analyze(messages[:i+1])
        metrics.append({
            'iteration': i,
            'message': messages[i],
            'analysis': analysis
        })
    
    return metrics

def plot_analysis(metrics, title):
    """Plot the analysis results."""
    # Create figure and axis objects
    fig, ax1 = plt.subplots(figsize=(12, 6))
    
    # Extract data
    iterations = [m['iteration'] for m in metrics]
    lexical_sim = [m['analysis'].lexical.similarity 
                   if m['analysis'].lexical is not None else None 
                   for m in metrics]
    semantic_sim = [m['analysis'].semantic.similarity
                    if m['analysis'].semantic is not None else None
                    for m in metrics]
    semantic_surp = [m['analysis'].surprise.semantic_surprise
                     if m['analysis'].surprise is not None else None
                     for m in metrics]
    
    # Remove None values
    valid_indices = [i for i, v in enumerate(lexical_sim) if v is not None]
    iterations = [iterations[i] for i in valid_indices]
    lexical_sim = [v for v in lexical_sim if v is not None]
    semantic_sim = [v for v in semantic_sim if v is not None]
    semantic_surp = [v for v in semantic_surp if v is not None]
    
    # Plot similarities
    ax1.set_xlabel('Message Index')
    ax1.set_ylabel('Similarity Score', color='tab:blue')
    ax1.plot(iterations, lexical_sim, label='Lexical Similarity',
             marker='o', color='tab:blue', alpha=0.6)
    ax1.plot(iterations, semantic_sim, label='Semantic Similarity',
             marker='s', color='tab:orange', alpha=0.6)
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.set_ylim(0, 1)
    
    # Plot surprise on second y-axis
    ax2 = ax1.twinx()
    ax2.set_ylabel('Surprise Score', color='tab:red')
    ax2.plot(iterations, semantic_surp, label='Semantic Surprise',
             marker='^', color='tab:red', alpha=0.6)
    ax2.tick_params(axis='y', labelcolor='tab:red')
    
    # Add legends
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    # Add grid
    ax1.grid(True, which='major', linestyle='-', alpha=0.5)
    ax1.grid(True, which='minor', linestyle=':', alpha=0.2)
    
    plt.title(title)
    plt.tight_layout()
    plt.show()
```
